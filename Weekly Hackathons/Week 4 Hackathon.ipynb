```python
cols = ['Total Savings', 'Total Food Expenditure', 'Total Non-Food Expenditure', 'Total Recreation Expenditure', 'Total Miscellaneous Expenditure', 'Main Source of Income']
corr_df = final_df[cols]
sns.pairplot(corr_df, hue='Main Source of Income')
plt.show()
```

```python
from scipy.stats import pearsonr
import matplotlib.pyplot as plt 

def corrfunc(x,y, ax=None, **kws):
    """Plot the correlation coefficient in the top left hand corner of a plot."""
    r, _ = pearsonr(x, y)
    ax = ax or plt.gca()
    # Unicode for lowercase rho (ρ)
    rho = '\u03C1'
    ax.annotate(f'{rho} = {r:.2f}', xy=(.9, .5), xycoords=ax.transAxes)
```

```python
g = sns.pairplot(corr_df)
g.map_lower(corrfunc)
plt.show()
```
